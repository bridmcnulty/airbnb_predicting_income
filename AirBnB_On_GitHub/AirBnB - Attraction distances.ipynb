```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
from math import sin, cos, sqrt, atan2, radians

df = pd.read_csv("/Users/bridmcnulty/Desktop/cleaned_data.csv")
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head(1)

# Find distance to London's top 12 attractions(Trip advisor's top 12 rated London attractions):
attrlocations = pd.read_csv("/Users/bridmcnulty/Desktop/attractions.csv", low_memory=False)
attrlocations.head()


df['latlongzipa'] = list(zip(df.latitude, df.longitude))
attrlocations['latlongzip'] = list(zip(attrlocations.latitude, attrlocations.longitude))

attrlocations.head(2)
```

```python
# Just do for British Museum:
distance_brm = []

for lat, long in df.latlongzipa:

    # approximate radius of earth in km
        R = 6373.0

        lat1 = radians(lat)
        lon1 = radians(long)
        lat2 = radians(attrlocations[attrlocations['landmark']=='british_museum'].latlongzip[0][0])
        lon2 = radians(attrlocations[attrlocations['landmark']=='british_museum'].latlongzip[0][1])

        dlon = lon2 - lon1
        dlat = lat2 - lat1

        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))

        distance = R * c  
        distance_brm.append(distance)
```
